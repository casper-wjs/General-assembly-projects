# Preprocessing

## Feature Engineering

Notice that we have many features such full_bath, half_bath and bsmt_full_bath, we can add them altogether to form 1 feature called total_baths

```python
train.shape
```

```python
train['total_baths'] = train['full_bath'] + train['half_bath'] + train['bsmt_full_bath'] 
```

Since we have a new feature of all the bathrooms together, we can drop the initial seperated bath columns

```python
train.shape
```

```python
train.drop(columns=['full_bath', 'bsmt_full_bath', 'half_bath'], inplace=True)
```

```python
train.shape
```

We will repeat the same for the test dataset.

```python
test.shape
```

```python
test['total_baths'] = test['full_bath'] + test['bsmt_full_bath'] + test['half_bath']
```

```python
test.drop(columns=['full_bath', 'bsmt_full_bath', 'half_bath'], inplace=True)
```

```python
test.shape
```

## Training Set and Validation Set

Let us break up our train dataset into a training set and a validation set.

```python
X_train = train.drop(columns='saleprice')
y_train = train['saleprice']
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.25, random_state=100)
```

```python
X_train.shape
```

```python
X_val.shape
```

```python
y_train.shape
```

```python
y_val.shape
```

We shall assign the test set as X_test and subsequently use this to predict saleprice (y_test) after training our model with the training and validation set.

```python
X_test = test
```

```python
X_test.shape
```

## Scaling

Let's scale our X_train, X_val and X_test data to ensure that our features are uniform in scale.

```python
ss = StandardScaler()
```

Note that we will would only have to fit the StandardScaler once and transform the data on the X_train, X_val and X_test.

```python
X_train_sc = ss.fit_transform(X_train)
X_val_sc = ss.transform(X_val)
X_test_sc = ss.transform(X_test)
```

# Modeling

## Linear Regression

```python
# Instantiate the model
lr = LinearRegression()
```

```python
# Fit the data to the model
lr.fit(X_train_sc, y_train)
```

Our train score without cross validation:

```python
# R^2 score
lr.score(X_train_sc, y_train)
```

By default, the train score is based on R^2 which is coefficient of determination. It is the proportion of variability in y we can explain with x. Our score means approximately 86% of the saleprice can be explained by our selected features.

We will now do cross validation to compare our scores on the train dataset.

```python
cross_val_score(lr, X_train_sc, y_train, cv=10).mean()
```

Linear Regression is not a good model and also not have good at addressing the issue of multicollinearity. Hence, this is shown when putting the model through a more rigorous process of cross val score.

We will also look at cross validating using root mean squared error as our metric.

```python
 np.sqrt(-cross_val_score(lr, X_train_sc, y_train, cv=10, 
                scoring='neg_mean_squared_error').mean())
```

The model is giving an extremely high error and is severely underfitting as the model is not complex enough. In reality, line of best fit is rarely linear. 

With cross val score, the data is partitioned differently in each fold and it is unable to score well on the all the test sets in each fold, resulting in very high RMSE.

### Evaluating with Root Mean Squared Error (RMSE) on Train and Validation Set

This Kaggle competition calls for the score to be based on RMSE.

We will use X_train_sc to predict saleprice and compare it against y_train data. We will then use RMSE to compare both results. This serves as a guide to how our model predicts saleprice to seen data (train dataset). It is likely that our model will perform slightly poorer to unseen data (test dataset).

```python
np.sqrt(metrics.mean_squared_error(y_train, lr.predict(X_train_sc)))
```

Similarly, we will evaluate our model based on the validation set.

```python
np.sqrt(metrics.mean_squared_error(y_val, lr.predict(X_val_sc)))
```

The goal is to get RMSE to be as close to 0 as possible. Strangely, RMSE is lower in the validation set which shouldn't be the case as training error almost always underestimate validation error.

We will have a look at the saleprice prediction based on the model.

```python
lr_pred = lr.predict(X_test_sc)
lr_pred
```

## Lasso Regression

```python
# Instantiate the model
lasso_cv = LassoCV(n_alphas=1000, cv=10, max_iter=500000)
```

```python
# Fit the training data to the model
lasso_cv.fit(X_train_sc, y_train)
```

```python
# Finding the optimal alpha
lasso_cv.alpha_
```

This will be our optimal alpha (hyperparamter) which we will put into Lasso model. This optimal alpha penalizes coefficients of features which are too far from zero. As a result, these coefficients are reduced to zeros under Lasso, which is also a form of feature selection.

We will cross validate the lasso R^2 with the optimal alpha.

```python
lasso = Lasso(alpha=lasso_cv.alpha_)
```

```python
lasso.fit(X_train_sc, y_train)
```

```python
lasso.score(X_train_sc, y_train)
```

We will look at the lasso coefficients to see which features had the most impact in predicting saleprice.

```python
df_lasso = pd.DataFrame({'features':X_train.columns,
                         'coeff':lasso.coef_})
df_lasso
```

We will also create another column for absolutes of the coefficients to account for highly negative correlation, if any.

```python
df_lasso['absolute_coeff'] = np.abs(df_lasso['coeff'])
```

```python
df_lasso.sort_values(by='absolute_coeff', ascending=False).head(10)
```

Based on Lasso Regression, the 5 most important features as a predictor of saleprice is:
1. overall_qual
2. gr_liv_area
3. neighborhood
4. kitchen_qual
5. 1st_flr_sf

We will also look at the bottom features which represent poor predictors.

We also note that there were features zeroed-out by Lasso. In this case, this is a form of feature selection as well. We will have look at features which were zero-out.

```python
df_lasso.loc[df_lasso['coeff'] == 0, :]
```

```python
df_lasso.loc[df_lasso['coeff'] == 0, :].shape
```

- There are 12 features which were zero-out by Lasso. 
- These features were computed by the model as features which did not have meaningful coefficients as predictor of the saleprice

We will also assess our model on the validation set.

```python
lasso.score(X_val_sc, y_val)
```

Lasso's score is marginally lower than Linear Regression's score.

### Evaluating with Root Mean Squared Error (RMSE) on Train and Validation Set

We will use X_train_sc to predict saleprice and compare it against y_train data. We will then use RMSE to compare both results.

```python
np.sqrt(metrics.mean_squared_error(y_train, lasso.predict(X_train_sc)))
```

Similarly, we will evaluate our model based on the validation set.

```python
np.sqrt(metrics.mean_squared_error(y_val, lasso.predict(X_val_sc)))
```

We will have a look at the saleprice prediction based on the model.

```python
lasso_pred = lasso.predict(X_test_sc)
lasso_pred
```

The goal is to get RMSE to be as close to 0 as possible. Compared to our Linear Regression model, Lasso performs more poorly? It could be due to a higher alpha hypermeter which may be 'overpunishing' the features?

## Ridge Regression

```python
# Instantiate the model
ridge_cv = RidgeCV(alphas = np.logspace(0, 5, 100),
                   cv = 10)
```

```python
# fit the model
ridge_cv.fit(X_train_sc, y_train)
```

```python
ridge_cv.alpha_
```

This will be our optimal alpha which we will put into Lasso model. This optimal alpha penalizes coefficients of features which are too far from zero, thus enforcing them to small. As a result, these coefficients are reduced to almost zero. This reduces the model complexity while keeping all the variables in the model. 

We will cross validate the ridge R^2 with the optimal alpha.

```python
ridge = Ridge(alpha=ridge_cv.alpha_)
```

```python
ridge.fit(X_train_sc, y_train)
```

```python
df_ridge = pd.DataFrame({'features':X_train.columns,
                         'coeff':ridge.coef_})
df_ridge
```

We will also create another column to show absolute coefficients for each features as we also want to see for highly negative correlations which will inversely affect the saleprice.

```python
df_ridge['absolute_coeff'] = np.abs(df_ridge['coeff'])
```

```python
df_ridge.sort_values(by='absolute_coeff', ascending=False).head(10)
```

Based on Ridge Regression, the 5 most important features as a predictor of saleprice is:
1. overall_qual
2. gr_liv_area
3. neighborhood
4. kitchen_qual
5. 1st_flr_sf

which are the same features for Lasso but interestingly in different orders. 

For example:
- gr_liv_area was ranked 2nd in Lasso but 3rd in Ridge
- neighborhood was ranked 3rd in Lasso but 4th in Ridge
- kitchen_qual was ranked 4th in Lasso but 5th in Ridge
- 1st_flr_sf was ranked 5th in Lasso but 2nd in Ridge

A possible explanation was perhaps how both models treat and penalizes features differently.

We will have a look at features which were poor predictors of saleprice.

```python
df_ridge.sort_values(by='absolute_coeff', ascending=True).head(10)
```

Key observations:
- year_built performed the poorest. This could be explained by referring to the data dictionary. There is also a feature called year remodel, meaning that aged houses could still have high saleprice if it was remodelled recently. We note that this was picked out by the Lasso and zeroed out as a feature.
